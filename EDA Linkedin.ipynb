```python
df["salary"] = df["salary"].fillna("Nan")
```

```python
start = []
end = []
for i in df["salary"]:
    if i == "Nan":
        start.append("Nan")
        end.append("Nan")
    else:
        a = i.split(".")
        res = re.sub(r'[^\w\s]', '', str(a))
        v = re.findall(r'\d+', str(res))
        start.append(v[0])
        end.append(v[2])
        
        
```

```python
df["Start_salary"] = start
df["End Salary"] = end
```

```python
df.drop(["salary"],axis=1,inplace=True)
```

```python
Seniority_level = []
Employment_type = []
Job_function = []
Industry = []
# ss = -1
for i in df["criteria"]:
    try:
        val_0 = i.split("}")[0]
        val_0 = val_0.split(":")
    #     ss = ss+1
    #     print(ss)
        val_0 = re.sub(r'[^\w\s]', '', str(val_0[1]))
        Seniority_level.append(val_0)
    except IndexError:
        Seniority_level.append("null")
        
    try:    
        val_1 = i.split("}")[1]
        val_1 = val_1.split(":")
        val_1 = re.sub(r'[^\w\s]', '', str(val_1[1]))
        Employment_type.append(val_1)
    except IndexError:
        Employment_type.append("null")
        
    try:    
        val_2 = i.split("}")[2]
        val_2 = val_2.split(":")
        val_2 = re.sub(r'[^\w\s]', '', str(val_2[1]))
        Job_function.append(val_2)
    except IndexError:
        Job_function.append("null")    
        
    try:
        val_3 = i.split("}")[3]
        val_3 = val_3.split(":")
        val_3 = re.sub(r'[^\w\s]', '', str(val_3[1]))
        Industry.append(val_3)
    except IndexError:
        Industry.append("null")   
```

```python
df["Seniority_level"] = Seniority_level
df["Employment_type"] = Employment_type
df["Job_function"] = Job_function
df["Industry"] = Industry 
```
